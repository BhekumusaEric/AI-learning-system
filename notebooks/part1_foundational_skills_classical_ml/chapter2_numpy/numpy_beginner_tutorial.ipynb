## Section 1 · What is NumPy?

NumPy stands for **Numerical Python**.

The core idea: instead of storing numbers in a Python list and looping over them
one at a time, NumPy stores them in an **array** and operates on all of them
simultaneously using fast C code under the hood.

Run the cell below to import it. The alias `np` is universal — every NumPy tutorial
and job you will ever encounter uses it.


````python
import numpy as np

print("NumPy version:", np.__version__)
print("Ready!")
````

## Section 2 · Your first array

Let's create a NumPy array from a Python list and compare them side by side.


````python
# A regular Python list
grades_list = [85, 90, 78, 92, 88]

# A NumPy array — same data, different container
grades_array = np.array([85, 90, 78, 92, 88])

print("Python list  :", grades_list)
print("NumPy array  :", grades_array)
print()
print("Type of list  :", type(grades_list))
print("Type of array :", type(grades_array))
````

````python
# The big difference — doing maths

grades_list  = [85, 90, 78, 92, 88]
grades_array = np.array([85, 90, 78, 92, 88])

# Add 5 bonus points to every grade

# With a list — you have to write a loop
bonus_list = [g + 5 for g in grades_list]

# With NumPy — just write the maths directly
bonus_array = grades_array + 5

print("List result  :", bonus_list)
print("Array result :", bonus_array)
print()
print("NumPy also works with: -  *  /  **  and more")
print("All of them apply to every element at once.")
````

## Section 3 · Different ways to create arrays

You will not always type your data by hand.
Here are the most common ways to create arrays.


````python
# From a list (you already know this one)
a = np.array([10, 20, 30, 40, 50])
print("From list       :", a)

# A range of numbers (like Python's range, but returns an array)
b = np.arange(0, 10, 2)   # start, stop, step
print("np.arange       :", b)

# Evenly spaced numbers between two values
c = np.linspace(0, 1, 6)  # start, stop, how many
print("np.linspace     :", c)

# An array of all zeros
d = np.zeros(5)
print("np.zeros        :", d)

# An array of all ones
e = np.ones(5)
print("np.ones         :", e)

# Random numbers between 0 and 1
np.random.seed(42)         # seed = same random numbers every time
f = np.random.rand(5)
print("np.random.rand  :", f.round(3))
````

## Section 4 · Shape and dimensions

Arrays can be 1D (a list of numbers), 2D (a table), or even higher.

In machine learning, your dataset is almost always a **2D array**:
- Rows = data points (patients, images, users...)
- Columns = features (measurements, pixel values, ratings...)

Understanding shape is critical. It tells you exactly what your data looks like.


````python
# 1D array — a single row of numbers
row = np.array([10, 20, 30, 40, 50])
print("1D array:", row)
print("Shape   :", row.shape)    # (5,)  — 5 elements, 1 dimension
print("Dims    :", row.ndim)
print()

# 2D array — rows and columns (like a spreadsheet)
table = np.array([
    [85, 90, 78],   # student 1: maths, science, english
    [92, 88, 95],   # student 2
    [70, 75, 80],   # student 3
    [60, 65, 55],   # student 4
])
print("2D array (4 students, 3 subjects):")
print(table)
print("Shape  :", table.shape)   # (4, 3) — 4 rows, 3 columns
print("Dims   :", table.ndim)
print("Total elements:", table.size)
````

````python
# Reshaping — changing the shape without changing the data
a = np.arange(12)
print("Original (1D):", a)
print("Shape:", a.shape)
print()

# Reshape to 3 rows, 4 columns
b = a.reshape(3, 4)
print("Reshaped to (3, 4):")
print(b)
print("Shape:", b.shape)
print()

# Tip: use -1 to let NumPy figure out one dimension automatically
c = a.reshape(6, -1)   # 6 rows, NumPy works out columns = 2
print("Reshaped to (6, -1):")
print(c)
print("Shape:", c.shape)
````

## Section 5 · Maths on arrays

This is where NumPy gets powerful.
Every basic operation works **element-wise** — it applies to every number at once.


````python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("a         :", a)
print("b         :", b)
print()
print("a + b     :", a + b)      # add element by element
print("b - a     :", b - a)      # subtract
print("a * b     :", a * b)      # multiply
print("b / a     :", b / a)      # divide
print("a ** 2    :", a ** 2)     # square every element
print("np.sqrt(a):", np.sqrt(a).round(3))  # square root of every element
````

````python
# Aggregate functions — summarise the whole array in one number
scores = np.array([72, 88, 91, 65, 79, 84, 95, 70])

print("Scores:", scores)
print()
print("Sum     :", scores.sum())
print("Mean    :", scores.mean())
print("Min     :", scores.min())
print("Max     :", scores.max())
print("Std dev :", scores.std().round(2))   # how spread out the scores are
print("Median  :", np.median(scores))
````

````python
# Working on rows or columns of a 2D array
# axis=0 means "go down the rows" (operate on each column)
# axis=1 means "go across the columns" (operate on each row)

marks = np.array([
    [85, 90, 78],   # student 1
    [92, 88, 95],   # student 2
    [70, 75, 80],   # student 3
    [60, 65, 55],   # student 4
])

print("Marks table:")
print(marks)
print()
print("Average per subject (axis=0, down columns):")
print(marks.mean(axis=0))   # one average per column
print()
print("Average per student (axis=1, across rows):")
print(marks.mean(axis=1))   # one average per row
````

## Section 6 · Indexing and slicing

Accessing specific elements or groups of elements from an array.

The syntax is similar to Python lists — but for 2D arrays, you use two indices.


````python
# 1D indexing
a = np.array([10, 20, 30, 40, 50])
#              0   1   2   3   4   <- index numbers

print("Array:", a)
print()
print("First element  a[0]  :", a[0])
print("Last element   a[-1] :", a[-1])
print("Third element  a[2]  :", a[2])
print()
print("First three    a[:3] :", a[:3])   # slice: start up to (not including) index 3
print("Last two       a[-2:]:", a[-2:])  # slice: last 2 elements
print("Every other    a[::2]:", a[::2])  # slice: step of 2
````

````python
# 2D indexing — [row, column]
marks = np.array([
    [85, 90, 78],   # row 0: student 1
    [92, 88, 95],   # row 1: student 2
    [70, 75, 80],   # row 2: student 3
    [60, 65, 55],   # row 3: student 4
])

print("Full table:")
print(marks)
print()
print("Student 2, subject 3    marks[1, 2] :", marks[1, 2])   # row 1, column 2
print("All of student 1        marks[0, :] :", marks[0, :])   # row 0, all columns
print("All maths scores        marks[:, 0] :", marks[:, 0])   # all rows, column 0
print()
print("Top 2 students, first 2 subjects:")
print(marks[:2, :2])
````

## Section 7 · Boolean masking

One of the most useful NumPy tricks.

A **boolean mask** is an array of True/False values.
When you use it to index an array, it gives you only the elements where the mask is True.

This is how you filter data without writing a loop.


````python
scores = np.array([72, 88, 91, 65, 79, 84, 95, 70])

# Step 1: create a mask — which scores are above 80?
mask = scores > 80
print("Scores:", scores)
print("Mask  :", mask)   # True where score > 80, False elsewhere
print()

# Step 2: use the mask to get just those scores
high_scores = scores[mask]
print("Scores above 80:", high_scores)
print()

# You can do it in one line
print("Scores below 75:", scores[scores < 75])
print("Scores equal 88:", scores[scores == 88])
````

````python
# Combining conditions with & (and) | (or)
scores = np.array([72, 88, 91, 65, 79, 84, 95, 70])

# Scores between 75 and 90
mid_range = scores[(scores >= 75) & (scores <= 90)]
print("Between 75 and 90:", mid_range)

# Scores below 70 or above 90
extremes = scores[(scores < 70) | (scores > 90)]
print("Below 70 or above 90:", extremes)

# Count how many students passed (>= 70)
passed = (scores >= 70).sum()   # True counts as 1, False as 0
print(f"\n{passed} out of {len(scores)} students passed")
````

## Section 8 · Useful functions to know

A quick tour of functions you will use often in data science and ML.


````python
# Sorting
scores = np.array([72, 88, 91, 65, 79, 84, 95, 70])

print("Original  :", scores)
print("Sorted    :", np.sort(scores))                    # smallest to largest
print("Reversed  :", np.sort(scores)[::-1])              # largest to smallest
print()

# argsort — gives you the INDEX positions that would sort the array
order = np.argsort(scores)
print("argsort   :", order)
print("Lowest scorer is student index:", order[0])
print("Highest scorer is student index:", order[-1])
````

````python
# Where — find indices where a condition is true
scores = np.array([72, 88, 91, 65, 79, 84, 95, 70])

failing = np.where(scores < 70)
print("Failing student indices:", failing[0])
print("Failing student scores :", scores[failing])
print()

# np.where with 3 arguments: if/else for arrays
# Replace scores: keep score if >= 70, otherwise replace with 0
result = np.where(scores >= 70, scores, 0)
print("Scores (0 if failed):", result)
````

````python
# Stacking arrays — combining data
# Think of this as adding rows or columns to a table

row1 = np.array([85, 90, 78])
row2 = np.array([92, 88, 95])
row3 = np.array([70, 75, 80])

# Stack vertically (add rows)
table = np.vstack([row1, row2, row3])
print("Stacked vertically (vstack):")
print(table)
print("Shape:", table.shape)
print()

col_extra = np.array([[95], [85], [88]])
table_with_extra = np.hstack([table, col_extra])
print("Added a column (hstack):")
print(table_with_extra)
print("Shape:", table_with_extra.shape)
````

## Section 9 · Real example — student grade analysis

Let's put everything together with a realistic dataset.

We have 10 students, each with scores in 4 subjects:
Maths, Science, English, History.

We will use NumPy to answer real questions about this data.


````python
import numpy as np

# 10 students x 4 subjects
# Columns: Maths, Science, English, History
np.random.seed(0)
grades = np.array([
    [78, 82, 91, 74],
    [95, 93, 88, 97],
    [60, 55, 70, 63],
    [88, 91, 85, 90],
    [72, 68, 75, 70],
    [85, 80, 78, 83],
    [50, 60, 65, 55],
    [91, 95, 89, 94],
    [67, 72, 80, 69],
    [83, 87, 82, 86],
])

subjects = ["Maths", "Science", "English", "History"]
print("Grade table (10 students x 4 subjects):")
print(grades)
print(f"\nShape: {grades.shape}")
````

````python
# Q1: What is each student's average grade?
student_avgs = grades.mean(axis=1)
print("Average grade per student:")
for i, avg in enumerate(student_avgs):
    print(f"  Student {i+1}: {avg:.1f}")
````

````python
# Q2: Which subject is hardest? (lowest class average)
subject_avgs = grades.mean(axis=0)
print("Average grade per subject:")
for subj, avg in zip(subjects, subject_avgs):
    print(f"  {subj:10s}: {avg:.1f}")

hardest = subjects[np.argmin(subject_avgs)]
print(f"\nHardest subject: {hardest} ({subject_avgs.min():.1f})")
````

````python
# Q3: How many students passed each subject? (pass mark = 70)
print("Students passing each subject (score >= 70):")
for subj, col in zip(subjects, grades.T):   # .T = transpose, loop over columns
    passed = (col >= 70).sum()
    print(f"  {subj:10s}: {passed}/10 students")
````

````python
# Q4: Who is the top student overall?
student_avgs = grades.mean(axis=1)
top_student  = np.argmax(student_avgs)

print(f"Top student   : Student {top_student + 1}")
print(f"Their average : {student_avgs[top_student]:.1f}")
print(f"Their grades  : {grades[top_student]}")
````

````python
# Q5: Which students need extra support? (average < 70)
student_avgs = grades.mean(axis=1)
struggling   = np.where(student_avgs < 70)[0]

print("Students needing extra support (avg < 70):")
if len(struggling) == 0:
    print("  None!")
else:
    for idx in struggling:
        print(f"  Student {idx+1}: average = {student_avgs[idx]:.1f}")
````

````python
# Q6: Normalise grades to 0-1 range
# This is called min-max scaling — used constantly in ML preprocessing

grade_min = grades.min()
grade_max = grades.max()

normalised = (grades - grade_min) / (grade_max - grade_min)

print("Original grades (first 3 students):")
print(grades[:3])
print()
print("Normalised grades (first 3 students) — now all between 0 and 1:")
print(normalised[:3].round(3))
print()
print(f"Original range : {grade_min} to {grade_max}")
print(f"Normalised range: {normalised.min():.2f} to {normalised.max():.2f}")
print()
print("This is exactly what StandardScaler and MinMaxScaler do in sklearn —")
print("they are just NumPy operations wrapped in a class.")
````
