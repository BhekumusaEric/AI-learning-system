````python
temps = np.array([22, 24, 27, 26, 21, 29, 23])

# 1. Median temperature
# HINT: np.median()
median_temp = ...
print("Median temp:", median_temp)

# 2. Number of days hotter than 25
# HINT: Create a condition, then use .sum()
hot_days_count = ...
print("Days > 25C:", hot_days_count)

# 3. Label as 'Hot' or 'Normal'
# HINT: np.where(condition, true_val, false_val)
labels = ...
print("Labels:", labels)
````

<details>
<summary><b>💡 Click to reveal the answer</b></summary>

```python
temps = np.array([22, 24, 27, 26, 21, 29, 23])

# 1. Median temperature
median_temp = np.median(temps)
print("Median temp:", median_temp)

# 2. Number of days hotter than 25
hot_days_count = (temps > 25).sum()
print("Days > 25C:", hot_days_count)

# 3. Label as 'Hot' or 'Normal'
labels = np.where(temps > 25, 'Hot', 'Normal')
print("Labels:", labels)
```
