# Dropout: Randomly Turn Off Neurons

Add `nn.Dropout` layers to an MLP to prevent overfitting.

- `dropout1`: p=0.5 after the first hidden layer
- `dropout2`: p=0.3 after the second hidden layer
- Never apply dropout to the output layer
- Always call `model.eval()` before evaluating — this disables dropout
