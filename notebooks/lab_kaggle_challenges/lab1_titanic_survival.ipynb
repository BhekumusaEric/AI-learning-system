### The Story
On April 15, 1912, the Titanic sank. 1,502 out of 2,224 passengers died. Some groups had better survival chances — women, children, and 1st class passengers. Can you build a model to predict who survived?

```python
# Step 1: Install and import everything we need
!pip install pandas scikit-learn seaborn matplotlib -q

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print('✅ All libraries loaded!')
```

```python
# Step 2: Load the Titanic dataset (built into seaborn!)
df = sns.load_dataset('titanic')
print('Dataset shape:', df.shape)
df.head()
```

```python
# Step 3: Explore the data
# How many survived vs died?
print('Survival counts:')
print(df['survived'].value_counts())
print()
print('Survival by gender:')
print(df.groupby('sex')['survived'].mean())
```

```python
# Step 4: Visualize survival rates
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

df.groupby('sex')['survived'].mean().plot(kind='bar', ax=axes[0], color=['#ff6b6b','#4ecdc4'], title='Survival by Gender')
df.groupby('pclass')['survived'].mean().plot(kind='bar', ax=axes[1], color=['gold','silver','#cd7f32'], title='Survival by Class')
df['age'].hist(bins=20, ax=axes[2], color='#00ff9d', alpha=0.7)
axes[2].set_title('Age Distribution')

plt.tight_layout()
plt.show()
```

```python
# Step 5: Prepare features for the model
# We need to:
# 1. Fill missing age values with the median age
# 2. Convert 'sex' from text to numbers (male=0, female=1)
# 3. Select our features

df_clean = df.copy()

# Fill missing ages with median
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())

# Convert sex to numbers
df_clean['sex_num'] = (df_clean['sex'] == 'female').astype(int)

# Select features: pclass, sex_num, age, sibsp (siblings/spouses), parch (parents/children)
features = ['pclass', 'sex_num', 'age', 'sibsp', 'parch']
X = df_clean[features]
y = df_clean['survived']

print('Features shape:', X.shape)
print('First 3 rows:')
print(X.head(3))
```
