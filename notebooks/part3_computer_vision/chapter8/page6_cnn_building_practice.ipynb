# CNN Building Practice

Build a Convolutional Neural Network to classify handwritten digits (MNIST).

**Architecture:**
- Conv2D(32, 3×3, ReLU)
- MaxPooling2D(2×2)
- Conv2D(64, 3×3, ReLU)
- MaxPooling2D(2×2)
- Flatten → Dense(128, ReLU) → Dropout(0.5) → Dense(10, softmax)

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
X_train = X_train.astype('float32') / 255.0
X_test  = X_test.astype('float32')  / 255.0
X_train = X_train.reshape(-1, 28, 28, 1)
X_test  = X_test.reshape(-1, 28, 28, 1)
y_train = keras.utils.to_categorical(y_train, 10)
y_test  = keras.utils.to_categorical(y_test,  10)

# 1. Build the CNN
model = keras.Sequential([
    # Conv layer: 32 filters, 3x3, ReLU
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    # MaxPooling 2x2
    layers.MaxPooling2D((2, 2)),
    # Conv layer: 64 filters, 3x3, ReLU
    
    # MaxPooling 2x2
    
    layers.Flatten(),
    # Dense 128, ReLU
    
    # Dropout 0.5
    
    # Output: 10 classes, softmax
    
])

# 2. Compile
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 3. Train (small subset for speed)
history = model.fit(X_train[:2000], y_train[:2000], epochs=3,
                    validation_data=(X_test[:500], y_test[:500]), verbose=1)

# 4. Evaluate
test_loss, test_accuracy = model.evaluate(X_test[:500], y_test[:500], verbose=0)
print(f'Test accuracy: {test_accuracy*100:.1f}%')
```

```python
assert model.count_params() > 1000, 'Model should have > 1000 parameters'
assert test_accuracy > 0.5, 'Should achieve > 50% accuracy'
assert len(history.history['accuracy']) == 3, 'Should train for 3 epochs'
print('✓ All tests passed!')
```
